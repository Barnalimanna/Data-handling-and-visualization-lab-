```python
#1.Create a rader chart showing skill levels of an employee in different areas.

import numpy as np
import matplotlib.pyplot as plt
labels=['Communication','Technical','Leadership','Teamwork','Problem-solving']
values=[8,9,6,7,8]
values+=values[:1]#reapeat first value to close the circle
angles=np.linspace(0,2*np.pi,len(values))

fig,ax=plt.subplots(figsize=(5,6),subplot_kw=dict(polar=True))
ax.plot(angles,values,'s--',linewidth=2)
ax.fill(angles,values,alpha=0.4)
ax.set_xticks(angles[:1])
ax.set_xticklabels(labels)
ax.set_title("Employee skill levels")
plt.show()
```

```python
#2.Create a rader chart showing skill levels of multiple employees in different areas.
import numpy as np
import matplotlib.pyplot as plt
labels=['Communication','Technical','Leadership','Teamwork','Problem-solving']

emp1=[8,9,5,7,3]
emp2=[7,9,5,4,8]
emp3=[6,7,6,5,9]
emp1+=emp1[:1]#reapeat first value to close the circle
emp2+=emp2[:1]
emp3+=emp3[:1]


angles=np.linspace(0,2*np.pi,len(labels),endpoint=False).tolist()
angles += angles[:1]  

fig,ax=plt.subplots(figsize=(5,6),subplot_kw=dict(polar=True))
ax.plot(angles,emp1,'s--',linewidth=2,label='emp1')
ax.plot(angles,emp2,'o-',linewidth=2,label='emp2')
ax.plot(angles,emp3,'o--',linewidth=2,label='emp3')
ax.fill(angles,emp1,alpha=0.4)
ax.fill(angles,emp2,alpha=0.5)
ax.fill(angles,emp3,alpha=0.4)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_title("Employee skill levels")
ax.legend()
plt.show()
```

```python
#4.Create a rader chart showing skill levels of multiple employees in different areas using for loop.
import numpy as np
import matplotlib.pyplot as plt

labels = ['Communication', 'Technical', 'Leadership', 'Teamwork', 'Problem-solving']
employees_data = {
    'Employee 1': [8, 9, 6, 7, 8],
    'Employee 2': [6, 10, 9, 6, 9],
    'Employee 3': [9, 8, 8, 8, 7]
}

for employee, skills in employees_data.items():
  employees_data[employee] = skills + skills[:1]


angles = np.linspace(0, 2 * np.pi, len(labels) + 1)

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for employee, skills in employees_data.items():
    ax.plot(angles, skills, 'o-', linewidth=2, label=employee)
    ax.fill(angles, skills, alpha=0.3)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_title("Employee Skill Levels")
ax.legend()
plt.show()

```
