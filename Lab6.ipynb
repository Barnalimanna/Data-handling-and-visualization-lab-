```python
#1.Create a multi-line chart to show placement trends for CSE,ECE,and ME from 2015 to 2024.
#Additionally, add a 5-year moving average to predict future trends.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
years=np.arange(2015,2025)
placements_cse=[100,120,130,140,160,180,190,200,220,240]
placements_ece=[80,100,110,115,130,150,160,170,185,200]
placements_me=[60,70,85,90,100,110,120,135,150,160]

df=pd.DataFrame({'Year':years,'CSE':placements_cse,'ECE':placements_ece,'ME':placements_me})
print(df)

df['CSE_MA']=df['CSE'].rolling(5).mean()
print(df['CSE_MA'])

df['ECE_MA']=df['ECE'].rolling(5).mean()
print(df['ECE_MA'])

df['ME_MA']=df['ME'].rolling(5).mean()
print(df['ME_MA'])

plt.plot(df['Year'],df['CSE'],label="CSE Placements",marker='o')
plt.plot(df['Year'],df['ECE'],label="ECE Placements",marker='s')
plt.plot(df['Year'],df['ME'],label="ME Placements",marker='d')

plt.plot(df['Year'],df['CSE_MA'],linestyle='--',label="CSE 5-yr Avg")
plt.plot(df['Year'],df['ECE_MA'],linestyle='-.',label="ECE 5-yr Avg")
plt.plot(df['Year'],df['ME_MA'],linestyle='-',label="ME 5-yr Avg")

plt.xlabel("Years")
plt.ylabel("Number of Placements")
plt.title("Placement Trends with Moving Averages")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2.Create a 3D scatter plot to analyze how CGPA,number of internship, and salary are related.
from mpl_toolkits.mplot3d import Axes3D

cgpa=np.random.uniform(6.0,10.0,100)
internships=np.random.randint(0,5,100)
salary=cgpa*internships*np.random.uniform(0.5,1.5,100)+np.random.uniform(1,2,100)

fig=plt.figure(figsize=(10,7))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(cgpa,internships,salary,c=salary,cmap='viridis')

ax.set_xlabel("CGPA")
ax.set_ylabel("Internships")
ax.set_zlabel("Salary(LPA)")
ax.set_title("3D Scatter: CGPA vs Internships vs Salary")
plt.show()
```

```python
#3.Create a violin plot to compare the distribution of salaries(LPA)across different CGPA ranges in CSE,ECE and ME.
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
cgpa=np.random.uniform(6.0,10.0,100)
salary=cgpa*np.random.uniform(0.8,1.5,100)+np.random.uniform(1,3,100)
branch=np.random.choice(['CSE','ECE','ME'],100)

df=pd.DataFrame({'CGPA':cgpa,'Salary':salary,'Branch':branch})

sns.violinplot(x="Branch",y="Salary",data=df,inner="quartile",palette="coolwarm")
plt.title("Salary Distribution by Branch")
plt.ylabel("Salary(LPA)")
plt.xlabel("Branch")
plt.grid(True)
plt.show()
```
