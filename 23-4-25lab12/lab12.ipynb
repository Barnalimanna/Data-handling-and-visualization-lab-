```python
#filling missing values with mode valu.
import numpy as np
import pandas as pd
data={
    'Name':['Alice','Bob','Charlie','David','Eve'],
    'Gender':['Female',np.nan,'Male',np.nan,'Female'],
    'City':['NY','LA',np.nan,'Chicago',np.nan],
    'Age':['25',np.nan,'30',np.nan,'40']
}
df=pd.DataFrame(data)

df_mode_filled=df.copy()
for column in ['Gender','City']:
    mode_val=df_mode_filled[column].mode()[0]
    df_mode_filled[column]=df_mode_filled[column].fillna(mode_val)
print("\nFilled with mode:")
print(df_mode_filled)
```

```python
#filling missing values with backward fill
df_bfill=df.fillna(method='bfill')
print("\nFilled with backward fill:")
print(df_bfill)
```

```python
#filling missing values with forward fill
df_ffill=df.fillna(method='ffill')
print("\nFilled with forward fill:")
print(df_ffill)
```
