```python
#1.How can we visualize the monthly revenue,expenses and profit of a company over six months using a connected dot plot and include labels and annotations for each data point, and present the profit separately as a bar char for comparison.
import matplotlib.pyplot as plt
months=['Jan','Feb','Mar','Apr','May','Jun']
revenue=[3000,3200,2900,3500,3700,3400]
expenses=[2500,2600,2300,2800,3000,2700]
profit=[rev-exp for rev,exp in zip(revenue,expenses)]
#connected dot plot for rev,exp and profit
plt.figure(figsize=(10, 6))
plt.plot(months,revenue,marker='s',linestyle='-',color='blue',label='Revenue')
plt.plot(months,expenses,marker='s',linestyle='-',color='red',label='Expenses')
plt.plot(months,profit,marker='s',linestyle='-',color='green',label='Profit')
#Annotate data points
for i, (m,r,e,p) in enumerate(zip(months,revenue,expenses,profit)):
    plt.annotate(f'R:{r}',(m,r),textcoords='offset points',xytext=(0,10),ha='center')
    plt.annotate(f'E:{e}',(m,e),textcoords='offset points',xytext=(0,10),ha='center')
    plt.annotate(f'P:{p}',(m,p),textcoords='offset points',xytext=(0,10),ha='center')
    
plt.xlabel('Months')
plt.ylabel('Amount')
plt.ylabel('Monthly Revenue,Expenses,and Profit')
plt.grid(True)
plt.legend()
#separate bar chart for profit
plt.figure(figsize=(8,4))
plt.bar(months,profit,color='green')
plt.xlabel('Months')
plt.ylabel('Profit')
plt.title('Monthly Profit')
plt.grid(axis='y')
plt.show()
```

```python
#2.How can we visualize employee distribution across different departments,showcasing the breakdown of employees 
#by job role within each department using a pictogram-like chart with stacked bar plots? Also,illustrate the totsl 
#employee count per department and provide clear labels,annotations and a legend to enhance readability.
import matplotlib.pyplot as plt
import numpy as np
departments=['HR','IT','Finance','Sales','Admin']
#Employee count per job role within each department
job_roles=['Managers','Analysts','Developers','Sales Reps','clerks']
employee_data={
    'HR':[3,2,0,0,5],
    'IT':[5,5,15,0,0],
    'Finance':[2,7,0,0,6],
    'Sales':[4,2,0,24,0],
    'Admin':[1,3,0,0,8],
}
fig, ax=plt.subplots(figsize=(10,6))
bottom= np.zeros(len(departments))
#print(bottom)
colors=['skyblue','lightcoral','lightgreen','gold','lightsalmon']

for i, role in enumerate(job_roles):
    values=[employee_data[dept][i] for dept in departments]
    ax.bar(departments,values,bottom=bottom,label=role,color=colors[i%len(colors)])
    bottom+=values
#Add data labels on top of each bar segment
for i, dept in enumerate(departments):
    total_employees= sum(employee_data[dept])
    ax.text(i,total_employees+1,str(total_employees),ha='center',va='bottom')
#customer the plot
ax.set_xlabel("Departments",fontsize=12)
ax.set_ylabel("Number of Employees",fontsize=12)
ax.set_title("Employee Distribution by Department and job role",fontsize=14)
ax.legend(title="Job Roles",loc="upper right",bbox_to_anchor=(1.2,1))#position the legend outside
plt.xticks(rotation=45,ha='right')#rotate overlapping elements
plt.tight_layout()#prevent overlapping elements
plt.show
```
