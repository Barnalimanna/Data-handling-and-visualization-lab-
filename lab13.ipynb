```python
#calculate the total number of null value present in data set

import pandas as pd

# Load your dataset
data = pd.read_csv('netflix_movies (1).csv')

# Calculate the total number of null values
total_null_values = data.isnull().sum().sum()
print(f"Total number of null values in the dataset: {total_null_values}")
```

```python
# visualize the missing value using heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Set the figure size


# Plot heatmap of missing values
sns.heatmap(data.isnull())
```

```python
#filling missing values with mode valu.
import numpy as np
import pandas as pd
data=pd.read_csv('netflix_movies (1).csv')
df=pd.DataFrame(data)

df_mode_filled=df.copy()
for column in ['director','cast','country']:
    mode_val=df_mode_filled[column].mode()[0]
    df_mode_filled[column]=df_mode_filled[column].fillna(mode_val)
print("\nFilled with mode:")
print(df_mode_filled)
```
